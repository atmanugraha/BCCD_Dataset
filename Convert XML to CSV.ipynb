```python
import xml.etree.ElementTree as ET
import csv
import os
```

```python
# Convert XMLs to CSV
xml_dir = 'BCCD/Annotations/'
xml_file = ''
csv_file = open('BCCD/train.csv','w')
csvwriter = csv.writer(csv_file)
csv_head = []
csv_head.append('image_name')
csv_head.append('cell_type')
csv_head.append('xmin')
csv_head.append('xmax')
csv_head.append('ymin')
csv_head.append('ymax')
csvwriter.writerow(csv_head)

for root,_,files in os.walk(xml_dir):
    all_files = files

# Total files: 364
# Train files: 328
# Test files: 36
train_files = all_files[:328]
test_files = all_files[328:]

for file in train_files:
    xml_file = os.path.join(xml_dir,file)
    tree = ET.parse(xml_file)
    root = tree.getroot()

    csv_content = []
    for obj in root.findall('object'):
        csv_row = []
        filename = root.find('filename')
        csv_row.append(filename.text)
        obj_name = obj.find('name')
        csv_row.append(obj_name.text)
        bndbox = obj.find('bndbox')
        xmin = bndbox.find('xmin')
        csv_row.append(xmin.text)        
        xmax = bndbox.find('xmax')
        csv_row.append(xmax.text)
        ymin = bndbox.find('ymin')
        csv_row.append(ymin.text)
        ymax = bndbox.find('ymax')
        csv_row.append(ymax.text)
        csv_content.append(csv_row)
    csvwriter.writerows(csv_content)

csv_file.close()

csv_file = open('BCCD/test.csv','w')
csvwriter = csv.writer(csv_file)
csvwriter.writerow(csv_head)
for file in test_files:
    xml_file = os.path.join(xml_dir,file)
    tree = ET.parse(xml_file)
    root = tree.getroot()

    csv_content = []
    for obj in root.findall('object'):
        csv_row = []
        filename = root.find('filename')
        csv_row.append(filename.text)
        obj_name = obj.find('name')
        csv_row.append(obj_name.text)
        bndbox = obj.find('bndbox')
        xmin = bndbox.find('xmin')
        csv_row.append(xmin.text)        
        xmax = bndbox.find('xmax')
        csv_row.append(xmax.text)
        ymin = bndbox.find('ymin')
        csv_row.append(ymin.text)
        ymax = bndbox.find('ymax')
        csv_row.append(ymax.text)
        csv_content.append(csv_row)
    csvwriter.writerows(csv_content)

csv_file.close()
```
